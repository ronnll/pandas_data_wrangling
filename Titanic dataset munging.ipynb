```python
##########################DATA WRANGLING / DATA MUNGING STARTING THE NEXT CELL#############################
```

```python
df[df.Fare.isnull()] # df.info() and we see that Fare has null value so let's fill it in
#df.groupby(['Pclass','Embarked'])['Fare'].median() <-- This gives you the groupby table, dont use this
medianfare = df.loc[(df.Pclass == 3) & (df.Embarked == 'S'),'Fare'].median() # Use this. This gives you VALUE of 8.05
df.Fare.fillna(medianfare, inplace= True) #Now it's filled in
```

```python
def GetTitle(name): # When we df.info(), we want to get value from name. Feature Engineer a title from name. (payment-identifier)
    firstnamewithtitle = name.split(',')[1] #take 'name' and splits by the comma. everything after the 'comma' is put here bc [1]
    title = firstnamewithtitle.split('.')[0]#takes the upper part and splits again by the '.' [0] gives you the title
    title = title.strip().lower() # strip removes all white spaces, and the .lower makes it all lower case.
    return title # return the title. Now the GetTitle function will give you the title for each Name
df['Title'] = df.Name.map(lambda x : GetTitle(x)) #This is how you add a feature in the pandas dataframe (INSERT)!!!!!!!

#df[df.Age.notnull()].boxplot('Age','Title'); #<-- you can see it graphed, and see that Title is a good candidate to predict Age
title_age_median = df.groupby('Title').Age.transform('median') #Get median Age for Title and put into variable, title_age_median
df.Age.fillna(title_age_median , inplace=True) #put title_age_median into all Nulls for Age.
```

```python
### OPTIONAL THIS IS DEALING WITH OUTLIERS. USING LOGS###
### OPTIONAL THIS IS DEALING WITH OUTLIERS. USING LOGS###
### OPTIONAL THIS IS DEALING WITH OUTLIERS. USING LOGS###

#OUTLIERS & TRANSFORMATIONS. Regarding Outliers, we can find them using Histogram, BoxPlot, or a Scatterplot Graph.
df.Age.plot(kind='hist', bins = 20, color = 'c'); #most age is in 20-30's, but few outliers in the 70's and 80's
df.loc[df.Age > 70] # Let's check out the rows where age is more than 70. Here we see 6 people
df.loc[df.Fare == df.Fare.max()] #Fare has a lot of outliers too when graphed, so let's seethe max fare

# Since we have a skewed dsitribution for Fare.We can apply a TRANSFORMATION to make it less skewed. Transformation is a common 
# way to treat outliers to some extent. Since we know passenger Fare will never be negative we can apply the Log Transformation.
Logfare = np.log(df.Fare + 1.0) # We add 1 to the fare to accomidate $0 Fares. This is bc log(0) is undefined.
```

```python
# FEATURE ENGINEERING [BINNING]. We cut the Fare into 4 bins, where each bin contains an EQUAL number of observations
df['Fare_Bin'] = pd.qcut(df.Fare, 4, labels=['very_low', 'low', 'high', 'very_high'])
```

```python
### OPTIONAL SELF MADE PRACTICE 
def selfmade(fare):
    if fare <= 100:
        return '<= 100'
    elif fare > 100 and fare <= 200:
        return '100 < fare <= 200'
    elif fare > 200 and fare <= 300:
        return '200 < fare <= 300'
    else:
        return '300 < fare'
    
df['SelfMade'] = df.Fare.map(lambda x : selfmade(x))
```

```python
#Feature Engineering Adult/Minor
df['AgeState'] = np.where(df['Age'] >= 18, 'Adult', 'Child') #If True then 'Adult', Else 'Child'
df['AgeState'].value_counts() # Now let's count them
pd.crosstab(df[df.Survived != -888].Survived, df[df.Survived != -888].AgeState) #crosstab survival between Adults and Chilren

#Checking if Bigger Family means higher survival rate. First we INSERT columns adding Parch(parents) + SibSp(Siblings) + 1(self)
df['FamilySize'] = df.Parch + df.SibSp + 1 
df['FamilySize'].plot(kind='hist', color='c'); # then we graph it to see the distribution. We see some Outliers
df.loc[df.FamilySize == df.FamilySize.max()] #Let's see what the max value of family size is.
df.loc[df.FamilySize == df.FamilySize.max(),['Name','Survived','FamilySize','Ticket']] # Same as above, but with SELECT
pd.crosstab(df[df.Survived != -888].Survived, df[df.Survived != -888].FamilySize) # Checking with all Familysizes

#Let's see if motherhood plays an impact on survival
df['IsMother'] = np.where(((df.Sex == 'female') & (df.Parch > 0) & (df.Age > 18) & (df.Title != 'Miss')),1,0) #If true 1, else 0
pd.crosstab(df[df.Survived != -888].Survived, df[df.Survived != -888].IsMother) # Here see see motherhood does affect survival

#Let's see if Cabin placement affects survival
df.Cabin.unique() # Shows all unique cabins
df.loc[df.Cabin == 'T'] # We extract all rows where Cabin value is 'T', and see that the T was prolly a typo cause only 1 value
df.loc[df.Cabin == 'T', 'Cabin'] = np.NaN #<-- run this to update T to Naan

def get_deck(cabin):
    return np.where(pd.notnull(cabin),str(cabin)[0].upper(),'z') #First letter of Cabin. If Null then 'Z'. We convert the cabin 
                                                        # value to a str() and pull the first letter of it, and capitalize it.

df['Deck'] = df['Cabin'].map(lambda x : get_deck(x)) # Map Cabin into get_deck using lambda and feature engineer that into 'Deck'
df.Deck.value_counts() #We see that Naan(Z) is most poular followed by C.
pd.crosstab(df[df.Survived != -888].Survived, df[df.Survived != -888].Deck) #We see that Deck is a good indicator of survival too
```

```python
# Binary Encoding [0,1 for Y,N]
# Label Encoding [Label Encoding 1,2,3 Low med high] 
# One hot Encoding [make the grid is_a,is_b, is_c]
df['Ismale'] = np.where(df.Sex == 'male', 1, 0) #INSERT 1 if male, else 0

df = pd.get_dummies(df,columns=['Deck', 'Pclass', 'Title', 'Fare_Bin', 'Embarked'])
# The get_dummies is VERY IMPORTANT. You can pass your cateogrical/numeric features and it gets inserted into df. If you print 
# -it will now show the one hot encoding. The get_dummies will create a feature for every possible value that exists. if 10 
# -titles, it will creat 10 features.
```

```python
# DROPPING COLUMNS IN PYTHON
df.drop(['AgeState'], axis=1, inplace=True) #Pass column name to remove from data. Axis=1 when dropping. inplace has to be true
```
